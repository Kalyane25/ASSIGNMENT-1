##  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset
n_samples = 100
x = np.random.rand(n_samples)
y = 2 * x + np.random.normal(0, 0.1, n_samples)  # y is correlated with x
z = -3 * x + np.random.normal(0, 0.5, n_samples)  # z is inversely correlated with x

# Create a DataFrame
data = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set seed for reproducibility
np.random.seed(30)

# Generate dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o', alpha=0.7)

# Set labels
ax.set_title('3D Scatter Plot of Data Points')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

## 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sales dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table for heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='g', linewidths=0.5)
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

## ## 4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(square root(x^2+y^2))

```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(df['X'].values.reshape(100, 100), 
                 df['Y'].values.reshape(100, 100), 
                 df['Z'].values.reshape(100, 100), 
                 cmap='viridis', edgecolor='none')

# Set labels and title
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

## Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(25)

# Generate dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population'] * 2,  # Bubble size proportional to population
            alpha=0.5, 
            color='blue')

# Annotate each bubble with the country name
for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=12, ha='right')

# Set labels and title
plt.title('Bubble Chart: Country Population vs. GDP')
plt.xlabel('GDP (in billion $)')
plt.ylabel('Population (in millions)')
plt.grid(True)

plt.show()
```

## .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values from 0 to 100
y = np.random.rand(n) * 100  # Random y values from 0 to 100
sizes = np.random.randint(10, 100, size=n)  # Random sizes for markers
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], size=n)  # Random colors

# Create a DataFrame for the data
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Create a Bokeh scatter plot
p = figure(title='Random Scatter Plot', x_axis_label='X Values', y_axis_label='Y Values', 
           width=800, height=400)  # Using width and height instead of plot_width and plot_height

# Add scatter points
p.scatter('x', 'y', size='sizes', color='colors', source=data, fill_alpha=0.6)

# Show the plot
show(p)

```

## 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import save
import pandas as pd

# Prepare the output for Jupyter Notebook
output_notebook()

# Create dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame for the data
data = pd.DataFrame({
    'fruits': fruits,
    'counts': counts
})

# Create a Bokeh bar chart
p = figure(x_range=data['fruits'], title='Fruit Counts', 
           x_axis_label='Fruits', y_axis_label='Counts', 
           height=400, width=600)

# Add bars to the figure
p.vbar(x='fruits', top='counts', source=data, width=0.9, 
        color='orange', legend_label='Counts')

# Set legend location
p.legend.location = 'top_left'

# Show the plot
show(p)
```

## 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure for the histogram
p = figure(title='Histogram of Random Data', 
           x_axis_label='Value', y_axis_label='Frequency', 
           height=400, width=600)

# Add bars to the figure (centering the bars)
p.vbar(x=edges[:-1], top=hist, width=0.7*(edges[1] - edges[0]), color='blue', alpha=0.6)

# Show the plot
show(p)
```

## 5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColumnDataSource

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)

# Define x and y values
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x_flat, 'y': y_flat, 'z': z_flat})

# Create a Bokeh figure for the heatmap
p = figure(title='Heatmap Example', 
           x_axis_label='X', y_axis_label='Y', 
           width=600, height=600)

# Create a color mapper
mapper = linear_cmap(field_name='z', palette=Viridis256, low=z_flat.min(), high=z_flat.max())

# Add rectangles for the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=mapper, line_color=None)

# Show the plot
show(p)

```
